```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import numpy as np
```

```python
# Create some data
x, y = np.meshgrid(np.arange(-20,20), np.arange(-20,20))
z = x*y
```

```python
cmap = colors.LinearSegmentedColormap.from_list("Precip", ["#00db16", "blue", "#d10000", 'black'])

plt.pcolormesh(x, y, z, cmap=cmap)
plt.colorbar()
```

```python
cmap = colors.LinearSegmentedColormap.from_list("Sunset", ["gold", 'lightskyblue', "navy"])

plt.pcolormesh(x, y, z, cmap=cmap)
plt.colorbar()
```
